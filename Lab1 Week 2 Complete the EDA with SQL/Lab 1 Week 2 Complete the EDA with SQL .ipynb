```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%sql select * from SPACEXTABLE;
```

```python
%sql select distinct Launch_Site from SPACEXTABLE;
```

```python
%sql select Launch_Site from SPACEXTABLE where Launch_Site like'CCA%' limit 5;
```

```python
%sql select sum(PAYLOAD_MASS__KG_) from SPACEXTABLE where Customer = 'NASA (CRS)';
```

```python
%sql select avg(PAYLOAD_MASS__KG_) from SPACEXTABLE where Booster_Version = 'F9 v1.1';
```

```python
%sql select min(Date) as min_date from SPACEXTABLE where Landing_Outcome='Success (ground pad)'  ;
```

```python
%sql select Booster_Version from SPACEXTABLE where (Landing_Outcome='Success (drone ship)') and (PAYLOAD_MASS__KG_>4000 and PAYLOAD_MASS__KG_<6000); 
```

```python
%sql select count(Landing_Outcome) as successful_mission from SPACEXTABLE where Landing_Outcome LIKE 'Success%';
```

```python
%sql select Landing_Outcome ,count(Landing_Outcome) as failure_mission from SPACEXTABLE where Landing_Outcome LIKE'Failure%';
```
